## Huggingface Spaces

Here we provide a sample notebook using Gradio, then show the steps to deploy it to Huggingface Spaces.

```python
from google.colab import userdata
import os
os.environ["OPENAI_API_KEY"] = userdata.get("OPENAI_API_KEY")
```

```python
%%writefile app.py
import gradio as gr
from openai import OpenAI
import os

client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

def ask_model(message):
    resp = client.responses.create(
        model="gpt-4.1-mini",
        input=message,
    )
    return resp.output_text


def chat_fn(message):
    return ask_model(message)

demo = gr.Interface(
    fn=chat_fn,
    inputs="text",
    outputs="text",
    title="My Super Cool Colab App"
)

if __name__ == "__main__":
    demo.launch(share=True)
```
