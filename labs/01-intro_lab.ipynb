## 🧭 Scientific Process

**Question:**  
How does changing the **system prompt** affect the model's responses to the same user question?

**Hypothesis:**  
I expect that different system prompts will change the tone, style, or content of the model's responses. For example, a "friendly teacher" prompt might produce more explanatory responses, while a "sarcastic robot" prompt might produce more humorous or brief responses.

**Experiment Plan:**  
1. Choose a fixed user prompt (e.g., "Explain this course to me in one or two sentences.").
2. Test the same user prompt with different system prompts (e.g., "friendly teacher", "sarcastic robot", "pirate").
3. Record the responses for each system prompt.
4. Compare the responses and note differences in tone, length, or content.
5. Try multiple user prompts to see if the pattern holds.

**Measurement (Metric):**  
- **Qualitative observations:** Tone (friendly, formal, humorous), style (brief, detailed), content differences
- **Quantitative (optional):** Response length (number of characters or words)

**Expected Shape:**  
Different system prompts should produce noticeably different responses. The model should adapt its tone and style to match the system prompt's instructions.

## 🔹 Part 3 — Testing Multiple Prompts

Now let's test your system prompt with multiple different user prompts to see if the pattern holds across different questions.


```python
# @title Try several prompts and store the results (list only)

prompts_to_try = EXAMPLE_LAB1_PROMPTS + [
    "Add your own prompt here (edit this string!)."
]

results = []

for p in prompts_to_try:
    reply = lab1_generate_reply(
        user_prompt=p,
        system_prompt=MY_SYSTEM_PROMPT,
        temperature=0.7,
    )
    print("==========")
    print("👤 Prompt:", p)
    print("🤖 Reply:", reply)
    # print("🤖 Reply:", reply[:200], "..." if len(reply) > 200 else "")
    results.append({"prompt": p, "reply": reply})

print(f"\nStored {len(results)} prompt/response pairs in `results`.")
```
