# Conditionals

Allow programs to make decisions.

````python
hero_name  = "Bizardo the Foolish!"
hero_hp = 100
hero_power = 10 # attack_power

villian_name = "Furgis the Wise"
villian_hp = 500
villian_power = 1

print(f"Hero: {hero_name}, HP: {hero_hp}, Power: {hero_power}")

print(f"Villian: {villian_name}, HP: {villian_hp}, Power: {villian_power}")
````

# Repetition

Computer excel at repeating mundane steps

Two basic types of loops:

- for loop
  - used when the number of repetions is known
- while loop
  - used when we don't know the number of repetitions to take

## while syntax

``` python
  while (condition):
     # instruction one
     # instruction two
     # ...
```



````python
print(f"Hero: {hero_name}, HP: {hero_hp}, Power: {hero_power}")

print(f"Villian: {villian_name}, HP: {villian_hp}, Power: {villian_power}")

while True:
  # hero attacks the villian
  villian_hp = villian_hp - hero_power
  # if villian hp <= 0, end loop
  if villian_hp <= 0:
    print(f"Stupidity wins, {hero_name} has prevailed!")
    break

  # villian attacks the hero
  hero_hp = hero_hp - villian_power

  # if hero hp <= 0, end loop
  if hero_hp <= 0:
    print(f"You failed, but society didn't, {villian_name} has prevailed")
    break
````

Functions add excess functionality to your code. This is nice because you can use this in mutiple areas.

````python
#basic health potion function

def hero_health_potion():
  #global modifies the outside variable
  global hero_hp
  #add to the heroes health
  hero_hp = hero_hp + 5
  #print
  print(f"Hero's HP has been increased to {hero_hp}")

#run the function
hero_health_potion()
````

````python
#Choose a combat move

def take_turn():
  #take in the globals
  global villian_hp
  global hero_power
  #ask for the inputs
  #make the input a integer, you can type different varients of 1 and make this work
  move = int(input("Press 1 to attack or 2 to take a potion: "))
  if move == 1:
    villian_hp = villian_hp - hero_power
    print(villian_hp)
  if move == 2:
    hero_health_potion()
  else:
    print("Invalid input")
````

````python
print(f"Hero: {hero_name}, HP: {hero_hp}, Power: {hero_power}")

print(f"Villian: {villian_name}, HP: {villian_hp}, Power: {villian_power}")

while True:
  # hero attacks the villian
  take_turn() # hero takes their turn
  # if villian hp <= 0, end loop
  if villian_hp <= 0:
    print(f"Stupidity wins, {hero_name} has prevailed!")
    break

  # villian attacks the hero
  hero_hp = hero_hp - villian_power

  # if hero hp <= 0, end loop
  if hero_hp <= 0:
    print(f"You failed, but society didn't, {villian_name} has prevailed")
    break
````
