```python
#try function
# 1. Success -> Wanted output -> Keep running
#2. Fail -> Give an exception
    # Using except -> except function -> print

#Example
try:
  result = 10/0
except:
  print("You can't divide by 0")

#10/0
```

```python
from logging import exception
#Try and except: user input
villian_hp = 10
hero_power = 1

def take_turn():
  #take in the globals
  global villian_hp
  global hero_power
  #ask for the inputs
  #make the input a integer, you can type different varients of 1 and make this work
  while True:
    try:
      #Dealing with a future error
      move = int(input("Press 1 to attack or 2 to take a potion: "))
      #no logical error, but still a wrong input
      if move != 1 and move != 2:
        print("Not in the moveset!")
        continue #try again
      break #break out of the loop
    except:
      print("Not a valid move!")

  if move == 1:
    villian_hp = villian_hp - hero_power
    print(villian_hp)
  if move == 2:
    print ("Added Health")
  else:
    print("Invalid input")

take_turn()
```
