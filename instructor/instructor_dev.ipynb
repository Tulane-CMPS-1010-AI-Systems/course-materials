```python
# @title 🔧 One-time setup: mount Drive and clone repo

drive.mount('/content/drive')

DEV_ROOT.mkdir(parents=True, exist_ok=True)

if COURSE_DIR.exists():
    print("Repo already exists at:", COURSE_DIR)
else:
    %cd {DEV_ROOT}
    !git clone {REPO_URL}
    print("Cloned repo to:", COURSE_DIR)

# set remote using your github token for auth
%cd {COURSE_DIR}
token = getpass.getpass("GitHub Personal Access Token: ")
owner = "Tulane-CMPS-1010-AI-Systems"
repo  = "course-materials"
remote_with_token = f"https://{token}@github.com/{owner}/{repo}.git"
!git remote set-url origin {remote_with_token}
!git remote -v  # verify

# set name and email
!git config --global user.email {EMAIL}
!git config --global user.name {NAME}
```
